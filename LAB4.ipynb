```python
Menu = []
total_price = 0

while True:
    menu_name = input("กรุณาการซื้อเมนู (หรือพิมพ์ exit เพื่อออก): ")
    if menu_name.lower() == "exit":
        break
    try:
        price = float(input(f"ราคาสินค้า {menu_name}: "))
        Menu.append({"name": menu_name, "price": price})
        total_price += price

        print(f"เพิ่ม {menu_name} ({price:.2f} บาท) เรียบร้อย!")
    except ValueError:
        print("กรุณาใส่ราคาเป็นตัวเลขเท่านั้น")

print("\n---- สรุปยอดออเดอร์ ----")

for index, item in enumerate(Menu, start=1):
    print(f"{index}. {item['name']} : {item['price']:.2f} บาท")

print(f"รวมทั้งหมด {len(Menu)} รายการ")
print(f"ราคารวม {total_price:.2f} บาท")
```
